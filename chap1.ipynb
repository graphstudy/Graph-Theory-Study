## 1.4 Isomorphic Graphs

```python
C6 = Graph ({"a":["b","c"] , "b":["a","d"], "c":["a","e"], "d":["b","f"], "e":["c","f"] , "f":["d","e" ]})
G1 = Graph ({1:[2 ,4] , 2:[1 ,3] , 3:[2 ,6] , 4:[1 ,5] , 5:[4 ,6] , 6:[3 ,5]})
G2 = Graph ({"a":["d","e"] , "b":["c","f"], "c":["b","f"], "d":["a","e"], "e":["a","d"] , "f":["b","c" ]})
```

```python
C6.is_isomorphic(G1)
```

```python
C6.is_isomorphic(G2)
```

```python
G1.is_isomorphic(G2)
```

### Permutation Equivalence

```python
C6.canonical_label().show()
G1.canonical_label().show()
G2.canonical_label().show()
```

```python
def check_graph_isomorphism(G1, G2):
    # Step 1: Find the canonical labels
    # Sage's "canonical_label" function does this for us.
    A1_prime = G1.canonical_label().adjacency_matrix()
    A2_prime = G2.canonical_label().adjacency_matrix()

    # Step 2: Check if the canonical labels are the same
    return A1_prime == A2_prime
```

```python
check_graph_isomorphism(C6, G1)
```

```python
check_graph_isomorphism(C6, G2)
```

```python
check_graph_isomorphism(G1, G2)
```
